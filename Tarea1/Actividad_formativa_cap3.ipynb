#### Actividad 1 con dataset data_studentlist.csv

```python
import pandas as pd

df = pd.read_csv('data_studentlist.csv', header='infer')

print("1) Altura promedio de los estudiantes varones: ")
print(df[df['gender'] == 'M']['height'].mean().round(2))

print("\n2) Altura promedio de las estudiantes mujeres: ")
print(df[df['gender'] == 'F']['height'].mean().round(2))

print("\n3) Peso promedio de los estudiantes varones: ")
print(df[df['gender'] == 'M']['weight'].mean().round(2))

print("\n4) Peso promedio de las estudiantes mujeres: ")
print(df[df['gender'] == 'F']['weight'].mean().round(2))

print("\n5) Estatura mayor entre los estudiantes varones: ")
print(df[df['gender'] == 'M']['height'].max())

print("\n6) Estatura menor entre las estudiantes mujeres: ")
print(df[df['gender'] == 'F']['height'].min())

print("\n7) El peso más ligero entre los estudiantes varones de estatura superior a 1.75 m: ")
print(df[(df['gender'] == 'M') & (df['height'] > 1.75)]['weight'].min())

print("\n8) El mayor peso entre las estudiantes de estatura inferior a 1.60 m:")
print(df[(df['gender'] == 'F') & (df['height'] < 1.60)]['weight'].max())

print("\n9) Nota promedio de alumnos sin ausencias ('N'):")
print(df[df['absence'] == 'N']['grade'].mean().round(2))

print("\n10) Nota promedio de alumnos con absentismo ('Y'):")
print(df[df['absence'] == 'Y']['grade'].mean().round(2))

print("\n11) Altura promedio de alumnos con grupo sanguíneo 'A' o 'AB':")
print(df[df['bloodtype'].isin(['A', 'AB'])]['height'].mean().round(2))

print("\n12) Altura promedio de estudiantes varones con sangre 'A' o 'AB':")
print(df[(df['gender'] == 'M') & (df['bloodtype'].isin(['A', 'AB']))]['height'].mean().round(2))

print("\n13) Edad promedio de alumnos con absentismo ('Y') y nota mayor o igual 3:")
print(df[(df['absence'] == 'Y') & (df['grade'] >= 3)]['age'].mean().round(2))
```

#### Actividad 2 con dataset data_census.csv

```python
import pandas as pd

df = pd.read_csv('data_census.csv', header='infer')

df['Population'] = df['Population'].str.replace(',', '').astype(int)
df['Households'] = df['Households'].str.replace(',','').astype(int)

print("1) Población total con ProvinceCode igual a 115 o 116: ")
print(df[df['Province Code'].isin([115, 116])]['Population'].sum())

print("\n2) Población promedio de las ciudades donde hay más hombres (GenderRatio > 1): ")
print(df[df['Gender Ratio'] > 1]['Population'].mean().round(2))

print("\n3) Lugares con más hombres (GenderRatio > 1) y menos de 2 personas por hogar: ")
print(df[(df['Gender Ratio'] > 1) & (df["Person's Household"] < 2)]['City Code'])

print("\n4) DataFrame ordenado en forma ascendente de 'Households' (muestra los 10 primeros): ")
df_ordenado=df.sort_values(by='Households', ascending=True)
df_ordenado.head(10)

```

#### Actividad 3 con dataset data_sales.csv

```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('data_sales.csv', header='infer')

print("1) Añade una nueva variable Amount = UnitPrice x Units y muestra las 5 primeras observaciones: ")
df['Amount'] = df['Units'] * df['Unit Price']
print(df.head(5))

print("\n2) Precio unitario promedio de cada región (groupby): ")
print(df.groupby('Region')['Unit Price'].mean().round(2))

print("\n3) Precio unitario promedio de cada región (pivot_table): ")
print(pd.pivot_table(df, values='Unit Price', index='Region', aggfunc='mean').round(2))

print("\n4) Precio unitario promedio para cada región en una línea de código (groupby): ")
print(df.groupby('Region')['Unit Price'].mean())

print("\n5) Precio unitario promedio y unidades para cada región en una línea (pivot_table): ")
print(pd.pivot_table(df, values=['Unit Price','Units'], index='Region', aggfunc='mean').round(2))

print("\n6) Total de unidades por región y tipo de artículo (pivot_table, faltantes=0): ")
print(pd.pivot_table(df, values='Units', index='Region', columns='Item', aggfunc='sum', fill_value=0))

print("\n7) Importe total de ventas por región y tipo de artículo (pivot_table, faltantes=0): ")
print(pd.pivot_table(df, values='Amount', index='Region', columns='Item', aggfunc='sum', fill_value=0, margins=True, margins_name='Total'))
```

#### Actividad 4 con dataset data_coffeeshop.csv

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('data_coffeeshop.csv', header='infer', na_values=[' '])
print(df.shape)
df.head(5)

# Inciso 1
print("\n1) Tabla de frecuencias de 'Year of Start' ordenada por año:\n")
freq_year = df['Year of Start'].value_counts().sort_index()
print(freq_year.head())

# Gráfica de línea 1997–2014
plt.figure(figsize=(7,5))
plt.plot(freq_year.index, freq_year.values, marker='o')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.title('Frecuencia anual de aperturas de tiendas (1997-2014)')
plt.xlim(1997, 2015)
plt.grid(True)
plt.show()

# Inciso 2
print("\n2) Tabla de frecuencias separada por estado actual ('In' vs 'Out'):\n")
df_in = df[df['Current State'] == 'In']
df_out = df[df['Current State'] == 'Out']
freq_in = df_in['Year of Start'].value_counts().sort_index()
freq_out = df_out['Year of Start'].value_counts().sort_index()

print("Frecuencias de tiendas 'In':\n")
print(freq_in.head())
print("\nFrecuencias de tiendas 'Out':\n")
print(freq_out.head())

# Gráficas
plt.figure(figsize=(7,5))
plt.plot(freq_in.index, freq_in.values, marker='o', label='In')
plt.plot(freq_out.index, freq_out.values, marker='o', label='Out')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.title('Tendencia anual por estado del negocio (1997-2014)')
plt.xlim(1997, 2015)
plt.grid(True)
plt.legend()
plt.show()
```

##### 3) Descripción general de la tendencia
Podemos observar que, a lo largo del periodo de 1997 a 2014, el número de tiendas que inician operaciones muestra una tendencia general de crecimiento. En los primeros años ambas curvas (“In” y “Out”) se mantienen cercanas, lo que indica que el número de aperturas y cierres era relativamente similar.

A partir de 2002 se aprecia un incremento más notorio en las tiendas que posteriormente cierran (“Out”), aunque la diferencia no es grande. Sin embargo, a partir de 2009 las curvas comienzan a separarse con mayor claridad y el número de tiendas que permanecen abiertas (“In”) aumenta de forma significativa, mientras que las tiendas que cierran muestran un crecimiento más moderado y posteriormente una disminución más o menos por 2013–2014.

Viéndola completa, la tendencia dice que el mercado se expandió de forma acelerada después de 2009 y que, aunque existen cierres, la mayoría de las tiendas abiertas tienden a mantenerse operando. Esto podría interpretarse como un fortalecimiento del sector y una mayor estabilidad en los establecimientos que continúan activos.
