# MANEJO DE TABLAS CON PANDAS

```python
import pandas as pd
import numpy as np 
import os
```

```python
df = pd.read_csv('data_studentlist.csv', header= 'infer') #Para nombrar las listas con el nombre de la primer fila
```

```python
df.head()
```

```python
df = df.rename(columns={
    "name": "NAME",
    "age": "AGE"
})
df.head ()
```

```python
df.columns = [col.upper() for col in df.columns]
df.head()
```

```python
df.columns = [col.lower() for col in df.columns]
df.head()
```

```python
df = df.add_prefix("prefix_")
df = df.add_suffix("_suf")

df.head()
```

```python
df.columns = ['NAME', 'GENDER' , 'AGE', 'GRADE', 'ABSENCE', 'BLOODTYPE', 'HEIGHT', 'WEIGHT']
df.head()
```

```python
df['IMC'] = 1000*df['WEIGHT']/df['HEIGHT']**2
df.head()
```

```python
df_left = df.loc[:, ['NAME','AGE', 'GENDER', 'ABSENCE']]
df_right = df.loc[:, ['NAME', 'BLOODTYPE', 'WEIGHT', 'HEIGHT']]
```

# Unir dataframes


```python
pd.merge(df_left,df_right, left_on="NAME", right_on="NAME", how="inner")

#left
#right
#outer
```

```python
df_left_peque = df_left.loc[:10,:]
df_right_peque = df_right.loc[7:,:]
```

```python
pd.concat([df_left_peque,df_right_peque], sort = True)
```

```python
pd.concat([df_left_peque,df_right_peque], axis=1, sort = True) #Al poner axis indica sobre cual va a operar
```

# Tabla pivote

Es una forma de visualizar ordenamente una tabla, reorganiza las tablas en una forma sencilla de entender

```python
my_dict = {"Size": ["L", "L", "M", "M", "M", "S", "S", "S", "S"],
           "Type": ["A", "A", "A", "B", "B", "A", "A", "B", "B"],
            "Location": ["L1", "L1", "L1", "L2", "L2", "L1", "L2", "L2", "L1"],
            "A": [1, 2, 2, 3, 3, 4, 5, 6, 7],
            "B": [2, 4, 5, 5, 6, 6, 8, 9, 9]}
df = pd.DataFrame(my_dict)
df
```

```python
dfr = pd.pivot_table(df, index=['Size','Type'], columns='Location',values='B') #  (dataframe, indices, columnas, valores)
dfr
```
