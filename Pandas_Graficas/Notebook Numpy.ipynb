# MANEJO DE TABLAS CON PANDAS

```python
import pandas as pd
import numpy as np 
import os
```

```python
df = pd.read_csv('data_studentlist.csv', header= 'infer') #Para nombrar las listas con el nombre de la primer fila
```

```python
df.head()
```

```python
df = df.rename(columns={
    "name": "NAME",
    "age": "AGE"
})
df.head ()
```

```python
df.columns = [col.upper() for col in df.columns]
df.head()
```

```python
df.columns = [col.lower() for col in df.columns]
df.head()
```

```python
df = df.add_prefix("prefix_")
df = df.add_suffix("_suf")

df.head()
```

```python
df.columns = ['NAME', 'GENDER' , 'AGE', 'GRADE', 'ABSENCE', 'BLOODTYPE', 'HEIGHT', 'WEIGHT']
df.head()
```

```python
df['IMC'] = 1000*df['WEIGHT']/df['HEIGHT']**2
df.head()
```

```python
df_left = df.loc[:, ['NAME','AGE', 'GENDER', 'ABSENCE']]
df_right = df.loc[:, ['NAME', 'BLOODTYPE', 'WEIGHT', 'HEIGHT']]
```

# Unir dataframes


```python
pd.merge(df_left,df_right, left_on="NAME", right_on="NAME", how="inner")

#left
#right
#outer
```
